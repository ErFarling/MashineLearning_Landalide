#空间分析:  从矢量数据提取特征    
1.计算点到水系，道路最小距离

```python
from shapely.ops import nearest_points
from shapely.wkt import loads
# 计算点到最近水系的距离
def get_distance(point, target_gdf):
    distances = []
    for geom in target_gdf.geometry:
        distances.append(point.distance(geom))
    return min(distances) if distances else None

# 为每个点计算到水系的距离
points_gdf['distance_to_water'] = points_gdf.geometry.apply(
    lambda x: get_distance(x, shuixi)
)

# 同理计算到道路的距离
points_gdf['distance_to_road'] = points_gdf.geometry.apply(
    lambda x: get_distance(x, daolu)
)
```

```python
# 空间连接：判断点是否落在岩性多边形内
rock_joined = gpd.sjoin(points_gdf, yanxing, how="left", predicate='intersects')

# 打印岩性数据（yanxing）的列名，确认实际字段名
print("岩性数据的列名：", yanxing.columns.tolist())
```

```python
import geopandas as gpd
import pandas as pd
# 查看数据结构和属性字段（确保包含地貌类型字段，如'TYPE'或'LAND_TYPE'）
print("dimao数据列：", dimao.columns)
print("前几行数据示例：")
print(dimao.head())
```

此处发现地貌数据中也存在LITHOLOGY（岩性）之前提过，这里选择合并或删除一个(避免冗余)     
SLOPE_MED（坡度中位数）RELIEF_MED（地形起伏度）ELEV_MED（高程中位数）#与之前不同此处都为区域性的均值，与之前的不构成重复

```python
import numpy as np
# 确保坐标系一致（重要！）
if not points_gdf.crs.equals(dimao.crs):
    points_gdf = points_gdf.to_crs(dimao.crs)

# 空间连接：将点与包含它的多边形关联
points_with_dimao = gpd.sjoin(
    points_gdf,
    dimao,
    how="left",
    predicate="within"  # 点必须位于多边形内部
)
```

```python
from sklearn.preprocessing import StandardScaler
# points_gdf已包含以下字段：
# 现有字段：NDVI, 宜黄DEM, 曲率, 坡度yihuangjiangshui, NEAR_FID, yihuangjiangshui
# 新提取字段：SLOPE_MED, RELIEF_MED, ELEV_MED, DOMSOIL, LANDFORM, SQKM, PERIMETER, AREA
# 提取关联后的地貌属性（如岩性、坡度等）
selected_features = [ 'SLOPE_MED','RELIEF_MED','ELEV_MED','DOMSOIL','LANDFORM','SQKM','PERIMETER','AREA']
points_gdf = points_with_dimao[selected_features + points_gdf.columns.tolist()].copy()
```

```python
# 1. 处理重复字段：计算局部与区域的差值
points_gdf.loc[:,'slope_diff'] = points_gdf['坡度'] - points_gdf['SLOPE_MED']
points_gdf.loc[:,'elevation_diff'] = points_gdf['宜黄dem'] - points_gdf['ELEV_MED']
```

```python
# 2. 处理DOMSOIL（土壤类型）#土壤类型独热编码
soil_dummies = pd.get_dummies(points_gdf['DOMSOIL'], prefix='soil')
points_gdf = pd.concat([points_gdf, soil_dummies], axis=1)
```

```python
# 3. 处理LANDFORM（地貌类型）地貌类型独热编码
landform_dummies = pd.get_dummies(points_gdf['LANDFORM'], prefix='landform')
points_gdf = pd.concat([points_gdf, landform_dummies], axis=1)
```

```python
# 4. 形状复杂度特征
points_gdf['shape_index'] = (points_gdf['PERIMETER'] ** 2) / (4 * np.pi * points_gdf['AREA'])#由周长面积计算
```
